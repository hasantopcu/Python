```python
from matplotlib import pyplot as plt
import pandas as pd
import numpy as np
```

```python
slices = np.random.randint(20,100, size=(5))
slices
```

```python
labels = ['Field 1','Field 2','Field 3','Field 4','Field 5']
colors = ['#FF4500','#008000','#DC143C','#20B2AA','#9400D3']
explode = [0 , 0 , 0 , 0 , 0.15]
```

```python
plt.title('Pie chart example')
plt.pie(slices , colors = colors , labels = labels  ,shadow=True, pctdistance = 0.5 , radius = 1.1 , 
        counterclock =False  ,   labeldistance=1.2 ,
        startangle=90 , autopct='%1.1f%%', explode = explode ,
        wedgeprops={'edgecolor':'green' , 'linewidth': 1})
plt.show()
```
