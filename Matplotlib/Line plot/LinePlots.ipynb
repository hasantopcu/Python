# I created a random dataset which named data.csv

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('data.csv')
```

```python
sub_data = data[['CalendarYear','CalendarQuarter','SalesAmount','Cost' , 'Quantity']]
```

```python
profit = (data['SalesAmount'].sub(data['Cost'])).mul(data['Quantity'])  
```

```python
result = pd.concat((sub_data , profit.rename('Profit')) , axis=1)
```

```python
a = result.groupby(['CalendarYear']).sum()['Profit']
plt.plot(a.keys() , a , label='Profit' , color='#FFA500' , marker='v')
plt.grid(True)
plt.tight_layout
plt.ylabel("Profit")
plt.xlabel('Years')
```

```python
result.groupby(['CalendarYear','CalendarQuarter']).sum()['Profit'].plot(marker='.')
plt.grid(True)
plt.tight_layout
plt.ylabel("Profit")
plt.xlabel('Years and Quartes')
```
