```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
from_host_istanbul = pd.read_csv('istanbul_sahibinden.csv')
print('Tuple counts for İstanbul: ' , len(from_host_istanbul))
from_host_ankara = pd.read_csv('ankara_sahibinden.csv')
print('Tuple counts for Ankara:   ' , len(from_host_ankara))
from_host_izmir =  pd.read_csv('izmir_sahibinden.csv')
print('Tuple counts for İzmir:    ' , len(from_host_izmir))
from_host_adana = pd.read_csv('adana_sahibinden.csv')
print('Tuple counts for Adana:    ' , len(from_host_adana))
```

```python
from_host_istanbul = from_host_istanbul[0:len(from_host_adana)]
from_host_ankara = from_host_ankara[0:len(from_host_adana)]
from_host_izmir = from_host_izmir[0:len(from_host_adana)]
```

```python
x = from_host_istanbul[['Fiyat','Net Metrekare']]
```

```python
indexes = []
for i in range(0 , len(x['Fiyat'])):
    x['Fiyat'][i] =  int(x['Fiyat'][i].replace('TL' , '').replace('.' , ''))
    if x['Net Metrekare'][i] =='Belirtilmemiş':
        indexes.append(i)
x = x.drop(indexes).reset_index(drop=True)
for j in range(0 , len(x['Fiyat'])):
    x['Net Metrekare'][j] =  int(x['Net Metrekare'][j].replace('m²' , '').replace('.' , ''))
    
```

```python
y = from_host_ankara[['Fiyat','Net Metrekare']]
```

```python
indexes = []
for i in range(0 , len(y['Fiyat'])):
    y['Fiyat'][i] =  int(y['Fiyat'][i].replace('TL' , '').replace('.' , ''))
    if y['Net Metrekare'][i] =='Belirtilmemiş':
        indexes.append(i)
y = y.drop(indexes).reset_index(drop=True)
for j in range(0 , len(y['Fiyat'])):
    y['Net Metrekare'][j] =  int(y['Net Metrekare'][j].replace('m²' , '').replace('.' , ''))
    if y['Fiyat'][j] > 4000:
            y['Fiyat'][j] = y['Fiyat'][j]/12
```

```python
plt.scatter(x['Net Metrekare'] , x['Fiyat'] , label='Istanbul' , color='green')
#
#plt.scatter(z['Net Metrekare'] , z['Fiyat']  , label='Izmir')
#plt.scatter(t['Net Metrekare'] , t['Fiyat']  , label='Adana')
plt.xlabel('Net square meter')
plt.ylabel('Monthly average rent(TL)')
plt.title('Monthly average rent(TL) by net square meter')
plt.ylim(0,20000)
plt.legend()
plt.grid(True)
plt.show()
```
